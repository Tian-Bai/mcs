##### Varying $n_{\textnormal{cal}}$

```python
import numpy as np
import pandas as pd

all_df = pd.DataFrame()

shape = 2

pd.set_option('display.max_columns', None)

all_df1 = pd.DataFrame()
for ncal in [100, 500, 3000]:
    df = pd.read_csv(f"vary_ncal\\DL_s, ntest=200, ntrain_f=6400, ncalib={ncal}, shape={shape}, seed=1_500, family=3.csv")
    df = df.groupby(['q'], as_index=False).mean()
    all_df1 = pd.concat((all_df1, df))
all_df = pd.concat((all_df, all_df1))

all_df
```
